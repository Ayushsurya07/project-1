```python
from PIL import Image

def find_min_max_pixel_values(image_path):
    # Open the image
    image = Image.open(image_path)

    # Convert image to grayscale for easier pixel value analysis
    grayscale_image = image.convert("L")

    # Get pixel values
    pixel_values = list(grayscale_image.getdata())

    # Find the minimum and maximum pixel values
    min_pixel_value = min(pixel_values)
    max_pixel_value = max(pixel_values)

    return min_pixel_value, max_pixel_value

# Example usage
image_path = '/Users/ayushsurya/Downloads/WhatsApp Image 2024-10-07 at 12.34.51.jpeg'  # Update with your image path
min_value, max_value = find_min_max_pixel_values(image_path)

print(f"Minimum pixel value: {min_value}")
print(f"Maximum pixel value: {max_value}")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def plot_image_histogram(image_path):
    # Open the image
    image = Image.open(image_path)

    # Convert image to grayscale for easier histogram representation
    grayscale_image = image.convert("L")

    # Get pixel values
    pixel_values = list(grayscale_image.getdata())

    # Create the histogram
    plt.hist(pixel_values, bins=256, range=(0, 255), color='black', alpha=0.7)

    # Add titles and labels
    plt.title("Histogram of Image")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")

    # Show the histogram
    plt.show()

# Example usage
image_path = '/Users/ayushsurya/Downloads/WhatsApp Image 2024-10-07 at 12.34.51.jpeg'  # Update with your image path
plot_image_histogram(image_path)
```

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

def apply_filters(image_path):
    # Open the original image
    original_image = Image.open(image_path)

    # Define a list of filters to apply
    filters = {
        "Original": original_image,
        "Blur": original_image.filter(ImageFilter.BLUR),
        "Contour": original_image.filter(ImageFilter.CONTOUR),
        "Detail": original_image.filter(ImageFilter.DETAIL),
        "Edge Enhance": original_image.filter(ImageFilter.EDGE_ENHANCE),
        "Emboss": original_image.filter(ImageFilter.EMBOSS),
        "Sharpen": original_image.filter(ImageFilter.SHARPEN),
        "Gaussian Blur": original_image.filter(ImageFilter.GaussianBlur(5)),  # Radius of 5
    }

    # Plot the images
    plt.figure(figsize=(12, 8))

    for i, (filter_name, image) in enumerate(filters.items()):
        plt.subplot(3, 3, i + 1)
        plt.imshow(image)
        plt.title(filter_name)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage
image_path = '/Users/ayushsurya/Downloads/WhatsApp Image 2024-10-07 at 12.34.51.jpeg'  # Update with your image path
apply_filters(image_path)
```
